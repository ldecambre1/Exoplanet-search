# **How Data Choices Affect Neural Network Performance**

In Q1 and Q2, we classified Kepler exoplanet candidates using traditional and neural network models on a fixed, imbalanced dataset.

Here, we explore how training data size and class balance impact neural network performance. Using the same features, preprocessing, architecture, and metrics as Q2, any performance differences reflect data choices alone.

We investigate two aspects:

* Training data size: Performance is evaluated on progressively smaller subsets of the dataset.
* Class balance: We compare the original imbalanced dataset with a rebalanced version (oversampling/undersampling).

Metrics focus on overall accuracy and minority-class predictions, allowing us to see how data quantity and balance affect model reliability.

## **Libraries**

```python
# PyTorch imports
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# Data handling and preprocessing
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Evaluation and visualisation
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

## **Loading and Preprocessing Data (same pipeline as Q2)**

We use the same dataset, features, and preprocessing steps as in Q2:
* Selected stellar and planetary parameters as input features.
* Target variable (koi_disposition) mapped to integers.
* Rows with missing values dropped.
* Features standardized using StandardScaler.
* Dataset split into training and test sets with stratified sampling to preserve class proportions.
* Converted to PyTorch tensors and loaded into DataLoader objects for training.

This ensures consistency with previous Questions, allowing us to isolate the impact of data quantity and class balance.

```python
# load dataset
raw_data = pd.read_csv("data/cumulative.csv")

# Display first few rows of dataset
raw_data.head()
```

```python
# Features to use for classification
features = [
    # STELLAR PARAMETERS
    "koi_steff",      # Stellar effective temperature (K)
    "koi_slogg",      # Stellar surface gravity (log10(cm/s^2))
    "koi_srad",       # Stellar radius (Solar radii)
    
    # PLANETARY CANDIDATE PARAMETERS
    "koi_period",     # Orbital period (days)
    "koi_duration",   # Transit duration (hours)
    "koi_depth",      # Transit depth (ppm)
    "koi_prad",       # Planetary radius (Earth radii)
]

# Target variable
target = 'koi_disposition'
```

```python
# Check for missing values in each column
missing_values = raw_data[features + [target]].isna().sum()
print("Missing values in each column:\n", missing_values)

# Calculate percentage of missing values
missing_percentage = (missing_values / len(raw_data)) * 100
print("\nPercentage of missing values in each column:\n", missing_percentage)

# Drop rows with missing values in features or target
cleaned_data = raw_data[features + [target]].dropna()
print(f"Rows after dropping NaNs:", len(cleaned_data))
```

```python
# Map classes to integers
target_mapping = {
    'CANDIDATE': 0,
    'CONFIRMED': 1,
    'FALSE POSITIVE': 2
}

# Apply mapping to target column
cleaned_data[target] = cleaned_data[target].replace(target_mapping)

# Drop any rows where target is still missing
cleaned_data = cleaned_data.dropna(subset=[target])
print("Rows after dropping remaining NaNs:", len(cleaned_data))

# Separate features (X) and target labels (y)
X = cleaned_data[features].values # Features as NumPy array
y = cleaned_data[target].values # Target labels as NumPy array

# Scale features (neural networks trains more efficiently with scaled input)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into training and test sets
# Stratify ensures class proportions are preserved in both sets
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size = 0.2,
    random_state = 42,
    stratify = y
)

# Convert NumPy arrays to PyTorch tensors for training
X_train_tensor = torch.tensor(X_train, dtype = torch.float32)
y_train_tensor = torch.tensor(y_train, dtype = torch.long)
X_test_tensor = torch.tensor(X_test, dtype = torch.float32)
y_test_tensor = torch.tensor(y_test, dtype = torch.long)

# Create PyTorch TensorDatasets
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

# Create PyTorch DataLoaders to efficiently feed data to neural network
train_loader = DataLoader(train_dataset, batch_size = 64, shuffle = True)
test_loader = DataLoader(test_dataset, batch_size = 64, shuffle = False)

# Print dataset sizes to verify split
print("Training samples:", len(train_dataset))
print("Testing samples:", len(test_dataset))
```

## **Baseline Neural Network (same architecture as Q2)**

We first train a neural network using the full training dataset with its natural class imbalance, replicating the architecture from Q2. This serves as our baseline, so any changes in performance in later experiments can be attributed to data modifications, not differences in the model.

The network is a simple feedforward model with one hidden layer and dropout for regularization. Training is performed for 20 epochs using the Adam optimizer, and performance is evaluated with accuracy, classification metrics, and a confusion matrix.

This baseline establishes a reference point for comparing how training data size and class balance affect model performance.

```python
# Define a neural network for exoplanet classification
class ExoplanetClassifier(nn.Module):
    def __init__(self):
        # Call the constructor of the parent nn.Module class
        super(ExoplanetClassifier, self).__init__()
        # Hidden layer takes 7 input features and outputs 64 values
        self.fc1 = nn.Linear(7, 64)
        # Output layer 3 neurons, one for each class (CANDIDATE, CONFIRMED, FALSE POSITIVE)
        self.fc2 = nn.Linear(64, 3)
        # ReLU activation function for hidden layer
        self.relu = nn.ReLU()
        # Dropout to prevent overfitting (here 20%)
        self.dropout = nn.Dropout(0.2)

    def forward(self, x):
        # Pass inputs through the first fully connected layer
        x = self.fc1(x)
        # Apply ReLU activation
        x = self.relu(x)
        # Apply dropout during training
        x = self.dropout(x)
        # Pass through the output layer
        x = self.fc2(x)

        return x
```

```python
# Initialise the neural network model
model = ExoplanetClassifier()

# Print the model architecture
print(model)
```

```python
# Training function for the neural network
def train_model(model, train_loader, criterion, optimizer, epochs = 20):
    # Set the model to training mode
    model.train()
    # Store loss values for each epoch
    loss_history = []
    # Loop over epochs
    for epoch in range(epochs):
        running_loss = 0.0
        # Loop over mini-batches of data
        for inputs, labels in train_loader:
            # Reset gradients from the previous iteration
            optimizer.zero_grad()
            # Forward pass: compute model predictions
            outputs = model(inputs)
            # Compute the loss between predictions and true labels
            loss = criterion(outputs, labels)
            # Backward pass: compute gradients of the loss
            loss.backward()
            # Update model parameters using the optimiser
            optimizer.step()
            # Accumulate batch loss
            running_loss += loss.item()

        # Compute average loss over all batches in this epoch
        epoch_loss = running_loss / len(train_loader)
        loss_history.append(epoch_loss)

        # Display training progress
        print(f"Epoch {epoch + 1}/{epochs}, Loss: {epoch_loss:.4f}")

    return loss_history
```

```python
# Train the model
loss_history = train_model(
    model, 
    train_loader, 
    nn.CrossEntropyLoss(),          # Loss function for multi-class classification
    optim.Adam(model.parameters(), lr = 0.001),  # Optimiser with learning rate 0.001
    epochs = 20
)
```

```python
# Plot training loss
plt.figure(figsize = (8,5))
plt.plot(range(1, len(loss_history)+1), loss_history, marker = 'o', color = 'b')
plt.title("Training Loss Over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import f1_score

def evaluate_model(model, test_loader):
    # Set the model to evaluation mode
    model.eval()
    all_preds = []   # store all predicted labels
    all_labels = []  # store all true labels
    correct = 0      # count of correct predictions
    total = 0        # total number of examples
    # Disable gradient computation for evaluation
    with torch.no_grad():
        for inputs, labels in test_loader:
            # Forward pass: compute model outputs
            outputs = model(inputs)
            # Choose class with highest predicted score
            _, predicted = torch.max(outputs, 1)
            # Store predictions and true labels
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            # Update counts for accuracy
            correct += (predicted == labels).sum().item()
            total += labels.size(0)

    # Compute overall accuracy
    accuracy = 100 * correct / total
    print(f"Accuracy: {accuracy:.2f}%\n")

    # Classification report
    print("Classification Report:")
    print(classification_report(
        all_labels, 
        all_preds, 
        target_names=['CANDIDATE', 'CONFIRMED', 'FALSE POSITIVE']
    ))

    # Confusion matrix heatmap
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize = (6,5))
    sns.heatmap(
        cm, 
        annot = True,
        fmt = 'd',
        cmap = 'Blues',
        xticklabels = ['CANDIDATE','CONFIRMED','FALSE POSITIVE'],
        yticklabels = ['CANDIDATE','CONFIRMED','FALSE POSITIVE']
    )
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

    f1_confirmed = f1_score(all_labels, all_preds, labels = [1], average = 'macro')
    return accuracy, f1_confirmed, all_preds
```

```python
# Evaluate the trained neural network on the test set
acc, f1, preds = evaluate_model(model, test_loader)
```

## **Effect of Training Data Size**

Neural networks generally perform better with more training data, but the improvement isnâ€™t always linear, and minority classes may benefit differently.

To explore this, we train the baseline model on progressively smaller fractions of the training set (10%, 25%, 50%, 75%, 100%) while keeping the test set fixed. Performance is evaluated using:

* Overall accuracy â€“ how well the model predicts all classes
* F1 score for confirmed planets â€“ to monitor performance on a minority class

This experiment helps us understand how sensitive the model is to the amount of training data and where diminishing returns begin.

```python
# Fractions of training data to explore
train_fractions = [0.1, 0.25, 0.5, 0.75, 0.99]

accuracy_results = []
f1_results = []

for frac in train_fractions:
    print(f"\nTraining on {int(frac*100)}% of the data")
    
    # Split a fraction of the training set (stratified)
    X_sub, _, y_sub, _ = train_test_split(
        X_train, y_train,
        train_size = frac,
        stratify = y_train,
        random_state = 42
    )

    # Convert to tensors and create DataLoader
    X_sub_tensor = torch.tensor(X_sub, dtype=torch.float32)
    y_sub_tensor = torch.tensor(y_sub, dtype=torch.long)
    sub_dataset = TensorDataset(X_sub_tensor, y_sub_tensor)
    sub_loader = DataLoader(sub_dataset, batch_size=64, shuffle=True)

    # Initialize fresh model for each fraction
    model = ExoplanetClassifier()
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    # Train
    train_model(model, sub_loader, criterion, optimizer, epochs=20)

    # Evaluate
    model.eval()
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Overall accuracy
    accuracy = 100 * (torch.tensor(all_preds) == torch.tensor(all_labels)).float().mean().item()
    accuracy_results.append(accuracy)

    # F1 for CONFIRMED class
    f1_confirmed = f1_score(all_labels, all_preds, labels=[1], average = 'macro', zero_division = 0)
    f1_results.append(f1_confirmed)

    # Classification report
    print(f"Accuracy: {accuracy:.3f}%")
    print(f"F1-score (Confirmed): {f1_confirmed:.3f}")
    print("Classification report:")
    print(classification_report(
        all_labels, 
        all_preds, 
        target_names = ['CANDIDATE', 'CONFIRMED', 'FALSE POSITIVE'],
        zero_division = 0
    ))
```

We investigated how varying the size of the training dataset affects the performance of our neural network model for classifying Kepler exoplanet candidates. The model was trained on progressively larger fractions of the training set: 10%, 25%, 50%, 75%, and 99%.

Key observations:
* Overall accuracy increases slightly as the training data size grows, from 56.1% (10%) to 61.8% (99%).
* Minority-class performance (CONFIRMED planets) also improves with more data, reflected in the F1-score rising from 0.416 to 0.585.
* Imbalanced classes: The model consistently predicts the majority class (FALSE POSITIVE) more accurately, while CANDIDATE planets remain the hardest to classify correctly.
* Diminishing returns: Accuracy and F1 improvements taper off beyond 50â€“75% of the training data, indicating that adding more examples has a smaller impact on performance.

**Conclusion**

Training data size has a noticeable effect on neural network performance, especially for minority classes, but there are diminishing returns once a substantial portion of the dataset is used. This highlights the importance of both dataset size and class balance in improving model reliability.

## **Effect of Class Imbalance**

The Kepler dataset is naturally imbalanced, with FALSE POSITIVE cases dominating the sample.

To investigate how class balance affects neural network performance, we compare:

* The original imbalanced dataset (baseline)
* A class-balanced dataset created via undersampling the majority classes

We will evaluate the impact on overall accuracy and F1 score for confirmed planets, and visualise performance using a confusion matrix.

```python
import numpy as np
import torch
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Scale features
X = cleaned_data[features].values
y = cleaned_data[target].values
X_scaled = scaler.transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size = 0.2, stratify = y, random_state = 42
)

# Identify indices for each class in training set
class_labels = np.unique(y_train)
indices_by_class = {label: np.where(y_train == label)[0] for label in class_labels}

# Find size of minority class
min_class_size = min(len(idx) for idx in indices_by_class.values())

# Undersample all classes to match minority
undersampled_indices = []
for idx in indices_by_class.values():
    sampled_idx = np.random.choice(idx, size=min_class_size, replace = False)
    undersampled_indices.append(sampled_idx)

# Combine undersampled indices
undersampled_indices = np.concatenate(undersampled_indices)

# Create balanced training data
X_train_bal = X_train[undersampled_indices]
y_train_bal = y_train[undersampled_indices]

# Convert to PyTorch tensors
X_train_tensor = torch.tensor(X_train_bal, dtype = torch.float32)
y_train_tensor = torch.tensor(y_train_bal, dtype = torch.long)
X_test_tensor = torch.tensor(X_test, dtype = torch.float32)
y_test_tensor = torch.tensor(y_test, dtype = torch.long)

# Create DataLoaders
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)
train_loader_bal = DataLoader(train_dataset, batch_size = 64, shuffle = True)
test_loader_bal = DataLoader(test_dataset, batch_size = 64, shuffle = False)

# Train the existing Q2 model
# Reuse model, criterion, optimizer from Q2
train_model(model, train_loader_bal, criterion, optimizer, epochs=20)

acc_bal, f1_bal, preds_bal = evaluate_model(model, test_loader_bal)
print("Balanced Dataset Performance")
print("Accuracy:", acc_bal)
print("F1 (Confirmed):", f1_bal)

cm = confusion_matrix(y_test, preds_bal)
sns.heatmap(
    cm,
    annot = True,
    fmt = "d",
    cmap = "Blues",
    xticklabels = ["Candidate", "Confirmed", "False Positive"],
    yticklabels = ["Candidate", "Confirmed", "False Positive"]
)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix (Balanced Dataset)")
plt.show()
```

**Observations**

After undersampling the majority classes to match the minority class size, we retrained the neural network on the balanced dataset.

Observations from the balanced dataset:

* **Overall accuracy:** 55.40%, which is slightly lower than the baseline model trained on the original imbalanced dataset (~61%). This drop in overall accuracy is expected since the model no longer benefits from the dominant FALSE POSITIVE class.
* **Minority-class performance (CONFIRMED planets):** F1-score increased to 0.586, compared with ~0.58 in the imbalanced dataset. Recall for CONFIRMED class improved (from 0.66 to 0.77), meaning the model correctly identifies more confirmed planets. Precision slightly decreased, reflecting more false positives in predicting this class.
* **Majority-class (FALSE POSITIVE) performance:** Precision dropped to 0.85 and recall to 0.53. The model is now less biased toward predicting the majority class, trading some majority-class accuracy for better minority-class detection.
* **CANDIDATE class:** Performance slightly improved (F1 = 0.36), showing more balanced predictions across all classes.

**Key takeaway:**
Balancing the dataset reduces overall accuracy but improves detection of minority classes. This demonstrates a trade-off between overall performance and fair representation of rare classes. In exoplanet classification, improving minority-class (confirmed planets) predictions can be more valuable than maximizing raw accuracy, especially when rare events are scientifically significant.
