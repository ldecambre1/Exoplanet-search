## **Libraries**

We begin by importing the Python libraries required to build, train, and evaluate a neural network classifier using PyTorch.

We will use:

* **PyTorch (`torch`)** to define the neural network architecture, loss function, and optimisation process.
* **scikit-learn** for splitting the dataset into training and test sets, feature standardisation, and evaluating model performance using classification metrics.
* **pandas** for loading and exploring the dataset.
* **matplotlib** and **seaborn** for visualising model performance, including confusion matrices.

Using scikit-learn for preprocessing and evaluation allows us to make a fair comparison with the traditional machine learning model used in Question 1.

```python
# PyTorch imports
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# Data handling and preprocessing
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Evaluation and visualisation
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

## **Loading the Dataset** 

We reload the Kepler exoplanet dataset using `pandas` to prepare it for neural network training. Most of the dataset exploration was already covered in `Question 1`, so we will only briefly revisit it in this notebook.


```python
# Reload dataset
raw_data = pd.read_csv("data/cumulative.csv")

# Display first few rows
raw_data.head()
```

## **Selecting Features and Target**

Next, we define which columns will be used as features.

We select features that describe:

* **Stellar parameters**: properties of the host star, which can influence transit measurements.  
* **Planetary candidate parameters**: properties of the transit itself, which are directly related to the likelihood of a planet being real.

The target variable is `koi_disposition`, which categorises each candidate as `CONFIRMED`, `CANDIDATE`, or `FALSE POSITIVE`.

Most of this feature selection was already covered in `Question 1`.

```python
# Features to use for classification
features = [
    # STELLAR PARAMETERS
    "koi_steff",      # Stellar effective temperature (K)
    "koi_slogg",      # Stellar surface gravity (log10(cm/s^2))
    "koi_srad",       # Stellar radius (Solar radii)
    
    # PLANETARY CANDIDATE PARAMETERS
    "koi_period",     # Orbital period (days)
    "koi_duration",   # Transit duration (hours)
    "koi_depth",      # Transit depth (ppm)
    "koi_prad",       # Planetary radius (Earth radii)
]

# Target variable
target = 'koi_disposition'
```

## **Handling Missing Values**

Neural networks cannot handle missing values (NaNs) directly, so we must address them before training.

In this notebook, we drop rows containing missing values in the selected features or target variable. This mirrors the preprocessing in Question 1 and ensures a fair comparison between the Decision Tree and Neural Network models.

While imputation techniques (e.g., mean or median imputation) are possible, only a small proportion of rows contain missing values, so dropping them is unlikely to significantly affect model performance.


```python
# Check for missing values in each column
missing_values = raw_data[features + [target]].isna().sum()
print("Missing values in each column:\n", missing_values)

# Calculate percentage of missing values
missing_percentage = (missing_values / len(raw_data)) * 100
print("\nPercentage of missing values in each column:\n", missing_percentage)

# Drop rows with missing values in features or target
cleaned_data = raw_data[features + [target]].dropna()
print(f"Rows after dropping NaNs:", len(cleaned_data))
```

## **Encoding the Target and Preparing Features**

Neural networks require numerical labels rather than strings. Here, we map the string classes in `koi_disposition` to integers as follows:

* `CANDIDATE`: 0  
* `CONFIRMED`: 1  
* `FALSE POSITIVE`: 2  

After encoding the target, we separate the dataset into features (`X`) and labels (`y`).  

We then standardise the features so that each has a mean of 0 and standard deviation of 1. Standardisation helps neural networks train more efficiently.  

Next, we split the dataset into training (80%) and test (20%) sets. We use stratification to ensure that the class proportions are preserved in both sets.  

Finally, we convert the NumPy arrays into PyTorch tensors and wrap them in `TensorDataset` objects. Using `DataLoader`s allows the neural network to process data efficiently in batches during training.  

This workflow ensures our data is fully prepared for training a neural network, while remaining consistent with the preprocessing applied in `Question 1`.

```python
# Map classes to integers
target_mapping = {
    'CANDIDATE': 0,
    'CONFIRMED': 1,
    'FALSE POSITIVE': 2
}

# Encode
cleaned_data[target] = cleaned_data[target].replace(target_mapping)

# Drop any rows where target is still missing
cleaned_data = cleaned_data.dropna(subset=[target])
print("Rows after dropping remaining NaNs:", len(cleaned_data))

# Separate features (X) and target labels (y)
X = cleaned_data[features].values # Features as NumPy array
y = cleaned_data[target].values # Target labels as NumPy array

# Scale features (neural networks trains more efficiently with scaled input)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into training and test sets
# Stratify ensures class proportions are preserved in both sets
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size = 0.2,
    random_state = 42,
    stratify = y
)

# Convert NumPy arrays to PyTorch tensors for training
X_train_tensor = torch.tensor(X_train, dtype = torch.float32)
y_train_tensor = torch.tensor(y_train, dtype = torch.long)
X_test_tensor = torch.tensor(X_test, dtype = torch.float32)
y_test_tensor = torch.tensor(y_test, dtype = torch.long)

# Create PyTorch TensorDatasets
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

# Create PyTorch DataLoaders to efficiently feed data to neural network
train_loader = DataLoader(train_dataset, batch_size = 64, shuffle = True)
test_loader = DataLoader(test_dataset, batch_size = 64, shuffle = False)

# Print dataset sizes to verify split
print("Training samples:", len(train_dataset))
print("Testing samples:", len(test_dataset))
```

## **Defining the Neural Network**

Next, we create a neural network to classify Kepler exoplanet candidates.  

Here's how it works:

* **Input layer**: The network takes 7 features describing the star and candidate planet.  
* **Hidden layers**: These layers allow the network to learn patterns in the data:  
  - First hidden layer: 128 neurons + ReLU activation  
  - Second hidden layer: 64 neurons + ReLU activation  
  - Third hidden layer: 32 neurons (extra depth to capture more complex patterns)  
* **Dropout**: After the first hidden layer, we randomly “turn off” 20% of neurons during training. This helps prevent overfitting, making the model generalise better to new data.  
* **Output layer**: 3 neurons, one for each class (`CANDIDATE`, `CONFIRMED`, `FALSE POSITIVE`).  

We use **ReLU** in the hidden layers to introduce non-linearity, which allows the network to learn complex relationships.  

The network is implemented using PyTorch by creating a subclass of `nn.Module` and defining a `forward` function, which specifies how data flows through the network.

```python
# Define a neural network for exoplanet classification
class ExoplanetClassifier(nn.Module):
    def __init__(self):
        super(ExoplanetClassifier, self).__init__()
        
        # Hidden layer takes 7 input features and outputs 64 values
        self.fc1 = nn.Linear(7, 64)
        # Output layer 3 neurons, one for each class (CANDIDATE, CONFIRMED, FALSE POSITIVE)
        self.fc2 = nn.Linear(64, 3)
        # ReLU activation function for hidden layer
        self.relu = nn.ReLU()
        # Dropout to prevent overfitting (here 20%)
        self.dropout = nn.Dropout(0.2)

    def forward(self, x):
        # Pass input through hidden layer and apply ReLU
        x = self.relu(self.fc1(x))
        # Apply dropout
        x = self.dropout(x)
        # Pass through output layer
        x = self.fc2(x)
        return x

# Initialise the model
model = ExoplanetClassifier()
print(model)
```

```python
# Training function for the neural network
def train_model(model, train_loader, criterion, optimizer, epochs = 30):
    # Put model in training mode
    model.train()
    loss_history = []

    # Loop over epochs
    for epoch in range(epochs):
        running_loss = 0.0

        # Loop over batches
        for inputs, labels in train_loader:
            # Zero previous gradients
            optimizer.zero_grad()
            # Forward pass
            outputs = model(inputs)
            # Compute loss between predictions
            loss = criterion(outputs, labels)
            # Backward pass
            loss.backward() 

            # Update model parameters
            optimizer.step()

            # Accumulate loss for this batch
            running_loss += loss.item()

        # Average loss over all batches in the epoch
        epoch_loss = running_loss / len(train_loader)
        loss_history.append(epoch_loss)

        # Print progress
        print(f"Epoch {epoch+1}/{epochs}, Loss: {epoch_loss:.4f}")

    return loss_history
```

```python
# Train the model
loss_history = train_model(
    model, 
    train_loader, 
    nn.CrossEntropyLoss(),          # Loss function for multi-class classification
    optim.Adam(model.parameters(), lr=0.001),  # Optimiser with learning rate 0.001
    epochs=20
)
```

```python
# Plot training loss
plt.figure(figsize=(8,5))
plt.plot(range(1, len(loss_history)+1), loss_history, marker='o', color='b')
plt.title("Training Loss Over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```

```python
def evaluate_model(model, test_loader):
    model.eval()  # evaluation mode
    all_preds = []
    all_labels = []
    correct = 0
    total = 0

    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            correct += (predicted == labels).sum().item()
            total += labels.size(0)

    # Accuracy
    accuracy = 100 * correct / total
    print(f"Accuracy: {accuracy:.2f}%")

    # Classification report
    from sklearn.metrics import classification_report
    print("Classification Report:")
    print(classification_report(all_labels, all_preds, target_names=['CANDIDATE','CONFIRMED','FALSE POSITIVE']))

    # Confusion matrix
    from sklearn.metrics import confusion_matrix
    import seaborn as sns
    import matplotlib.pyplot as plt

    cm = confusion_matrix(all_labels, all_preds)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['CANDIDATE','CONFIRMED','FALSE POSITIVE'],
                yticklabels=['CANDIDATE','CONFIRMED','FALSE POSITIVE'])
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

# Evaluate your trained NN
evaluate_model(model, test_loader)
```
